```python
qvi["YEAR&MONTH"] = pd.to_datetime(qvi["DATE"]).dt.to_period("M")
```

```python

qvi["YEAR&MONTH"]
```

```python
month_metrics = qvi.groupby(["STORE_NBR","YEAR&MONTH"]).agg({"TOT_SALES": "sum","LYLTY_CARD_NBR": "nunique",
                                                            "TXN_ID" : "nunique", "PROD_QTY" : "sum"})
month_metrics["PRICE_PERUNIT"] = month_metrics["TOT_SALES"]/month_metrics["PROD_QTY"]
month_metrics["CHIP_PER_TRANS"]=month_metrics["PROD_QTY"]/month_metrics["TXN_ID"]
month_metrics["TRANS_PERCUST"] = month_metrics["TXN_ID"]/month_metrics["LYLTY_CARD_NBR"]
month_metrics.drop(["TXN_ID"],axis = 1, inplace = True)
```

```python
lst = []
for i in month_metrics.index:
    if (i[1] >= 2019-02):
        if (i[1] <= 2019-04):
            lst.append(month_metrics.loc[i])
        month_metrics.drop(i, inplace = True)
lst = pd.DataFrame(lst)
```

```python
def calc_corr(store):
    a=[]
    month_metrix=month_metrics[['TOT_SALES','LYLTY_CARD_NBR']]

    for i in month_metrix.index:
        a.append(month_metrix.loc[store].corrwith(month_metrix.loc[i[0]]))
    qvi= pd.DataFrame(a)
    qvi.index=month_metrix.index
    qvi=qvi.drop_duplicates()
    qvi.index=[s[0] for s in qvi.index]
    qvi.index.name="STORE_NBR"
    return qvi
```

```python
def standard(qvi):
    qvi=qvi.abs()
    qvi['MAGNITUDE']=qvi.mean(axis=1)
    return qvi
```
